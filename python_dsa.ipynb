# String Methods

```
* Slicing - [start:stop(exclusive):step]

* Find/Change Characters
    * Character Case
        * .upper() and .lower() - change the case of the entire string
        * .capitalize() - Only first character capitalized and every other character in a string lowercased
        * .isupper() and .islower() - checks if all characters in the string are uppercase or lowercase
        * .swapcase() - change case of all

    * Character Types
        * isalpha() - alphabetic and at least one character
        * isnumeric() - numeric and at least one character
        * isalnum() - alphanumeric [a-zA-Z0-9] and at least one character
    
    * Counting characters
        * .count(substring) - the number of non-overlapping occurances of substring in the string
    
    * Find character index
        .find() and .rfind() return the lowest index in the string where substring occurs, staring from the left/right side
        
    * Check prefix/suffix
        .startswith() and .endswith() - case-sensitive

    * .strip(), lstrip(), rstrip() - remove leading/trailing whitespace
        * can also take characters to split too: "www.hi.com".strip(".cmow") -> "hi"
        * removeprefix() and removesuffix() for exact matching

* Process String
    * .split(' ', maxsplit=1) and .rsplit(' ', maxsplit=1)
        * obtain tokens of a string with a specified delimiter
        * split is for maxsplit starting left side and rsplit is for right side
    
    * .partition(sep)
        * return a three-tuple of (left, sep, right)
        * (string, '', '') if sep not found
    
    * .center(length, char), .ljust(length, char), .rjust(length, char)
        * Return a string of size legnth with string surrounded by char
        * center puts string in center, ljust puts string left, rjust puts string right
    
    * .zfill(length)
        * made a string of size length left-padded with zeros
        * a + or - prefix in string will be included, padded 0s start after

* Create String
    * ' '.join(list_of_strings)
        * concatenate the strings in the iterable with the string as delimiter
```

# List Methods

* slicing - arr[start:stop(exclusive):step]
    * [::-1] is reverse
* .append(val) - add val to end of list
* .clear() - remove all elements
* .copy() - return copy
* .count(val) - number of elements equal to val
* .extend(list2) - add list2 to end of list
* .index(val) - returns the index of the first element equal to val
* .insert(pos, val) - adds val at pos
* .pop(pos) - removes and returns element at pos
* .remove(val) - removes first element equal to val - returns ValueError if not found
* .reverse() - reverse list
* .sort() - sort list

````python
class Node:
    def __init__(self, val, next = None):
        self.val = val
        self.next = next
````

# Binary Tree

````python
class Node:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
````

# Queue

````python
queue = []

# enqueue()
queue.append(1)

# dequeue()
queue.pop(0)

# front()
queue[0]

# back()
queue[-1]
````

# Heap

# import heapq
.heapify(x) - transform x into a heap, in-place

.heappush(heap, item) - push item onto heap
.heappop(heap) - pop and return smallest item from heap
.heappushpop(heap, item) - push item, pop smallest item
.heapreplace(heap, item) - pop smallest item, push new item

.merge(*iterables, key, reverse) - merge multiple sorted inputs into one sorted iterable
.nlargest(n, iter, key) - return a list with the n largest elements from iter
.nsmallest(n, iter, key) - return a list with the n smallest elements from iter
